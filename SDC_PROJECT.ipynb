```python
from matplotlib import pyplot as plt
import numpy as np
import math
x = np.arange(0,math.pi*2,0.05)
#x = np.arange(0,10,1)
y = 2*np.sin(x)
print(x)
print(y)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
categories=["lion" ,"tiger", "horse"]
values1=[15, 24, 30]
values2=[20, 18, 25]
bar_width=0.35
bar_positions1=np.arange(len(categories))
bar_positions2=bar_positions1+ bar_width
plt.bar(bar_positions1 , values1 , width= bar_width , label="2021", color="blue")
plt.bar(bar_positions2 , values2 , width= bar_width , label="2022", color="skyblue")


plt.xlabel("categories")
plt.ylabel("values")
plt.title("grouped bar graph")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
categories=["transport" ,"income tax", "housing"]
values1=[15, 24, 30]
values2=[20, 18, 25]
plt.bar(categories, values1, label="First half", color="GREY")
plt.bar(categories, values2, bottom=values1, label="Second half", color="YELLOW")
plt.xlabel("categories")
plt.ylabel("values")
plt.title("static bar graph")
plt.legend()
plt.show()
```
